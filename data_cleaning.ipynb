```python
import pandas as pd

df = pd.read_csv("Uncleaned_DS_Jobs.csv")
df
```

Extract Degree/Qualifications. Look for trigger words like Master's, Bachelor's, Associate's Degree

```python
temp_df = df
```

```python
import re


dict = {
    "Phd": ["Doctorate", "Phd", "PHD", "PHd", "Doctorate's"],
    "Bachelor": ["Bachelor", "B.S.", "BS", "BS.c", "Bachelor's","Bachelors"],
    "Master": ["Master", "M.S.","MS","MS.c", "Master's", "Masters"]
}
degrees = []
for string in temp_df['Job Description']:

    degrees_mentioned = []
    for key, values in dict.items():
        for i in values:
            pattern = re.compile(fr'\b{re.escape(i)}\b')
            x = re.findall(pattern, string)
            if x:
                if key not in degrees_mentioned:
                    degrees_mentioned.append(key)
    degrees.append(degrees_mentioned)
    
```

```python
degrees = [ ', '.join(i) for i in degrees]
```

```python
temp_df['Degrees Mentioned'] = degrees
temp_df.columns
```

```python
# extract the average salary
import numpy as np

avg_salaries = []

for x in temp_df["Salary Estimate"]:
    text = x.split('(')[0].strip()
    nums = [ ((int(i))*1000) for i in re.findall(r'\d+', text)] 
    avg_salaries.append(np.average(nums))
```

```python
temp_df["average_salary"] = avg_salaries
temp_df = temp_df.drop(["Salary Estimate"], axis=1)
```

```python
import us

states = []

for x in temp_df["Location"]:
    if ',' in x:
        text = x.split(',')[1].strip()
        if us.states.lookup(text):
            state = us.states.lookup(text)
            states.append(state.name)
        else:      
            if text.strip() == "DC":
                states.append("Washngton,D.C.")
            else:
                text = x.split(',')[2].strip()
                if us.states.lookup(text):
                    state = us.states.lookup(text)
                    states.append(state.name)
    else:     
        if us.states.lookup(x):
            state = us.states.lookup(x)
            states.append(state.name)
        else:
            if x.strip() == "DC":
                states.append("Washngton,D.C.")
            else:
                states.append(x)
```

```python
temp_df['State'] = states
```

```python
# look for key words - skills

import re

counter = 0
skills = ["Python", "aws","tableau", "excel", "microsoft excel", "ms excel", "power bi", "powerbi", "sql","mysql", "machine learning"]

skills_col = []
for string in temp_df['Job Description']:
    skills_mentioned = []
    for i in skills:
        pattern = re.compile(fr'\b{re.escape(i)}\b', re.IGNORECASE)
        x = re.findall(pattern, string)
        if x:
            if i not in skills_mentioned:
                if i == "powerbi":
                    skills_mentioned.append("power bi")
                elif i == "microsoft excel" or i == "ms excel":
                    skills_mentioned.append("excel")
                else:    
                    skills_mentioned.append(i)
    skills_col.append(skills_mentioned)
    counter += 1
```

```python
temp_df['Python'] = None
temp_df['AWS'] = None
temp_df['Tableau'] = None
temp_df['Excel'] = None
temp_df['Power BI'] = None
temp_df['SQL'] = None
temp_df['MySQL'] = None
temp_df['Machine Learning'] = None
```

```python

i = 0
for x in temp_df['Python']:
    if "Python" in skills_col[i]:
        temp_df['Python'][i] = 1
    else:
         temp_df['Python'][i] = 0
    i += 1
i = 0
for x in temp_df['AWS']:
    if "aws" in skills_col[i]:
        temp_df['AWS'][i] = 1
    else:
         temp_df['AWS'][i] = 0
    i += 1
i = 0
for x in temp_df['Tableau']:
    if "tableau" in skills_col[i]:
        temp_df['Tableau'][i] = 1
    else:
         temp_df['Tableau'][i] = 0
    i += 1
i = 0
for x in temp_df['Excel']:
    if "excel" in skills_col[i]:
        temp_df['Excel'][i] = 1
    else:
         temp_df['Excel'][i] = 0
    i += 1
i = 0
for x in temp_df['Power BI']:
    if "power bi" in skills_col[i]:
        temp_df['Power BI'][i] = 1
    else:
         temp_df['Power BI'][i] = 0
    i += 1
i = 0
for x in temp_df['SQL']:
    if "sql" in skills_col[i]:
        temp_df['SQL'][i] = 1
    else:
         temp_df['SQL'][i] = 0
    i += 1

i = 0
for x in temp_df['MySQL']:
    if "mysql" in skills_col[i]:
        temp_df['MySQL'][i] = 1
    else:
         temp_df['MySQL'][i] = 0
    i += 1
i = 0
for x in temp_df['Machine Learning']:
    if "machine learning" in skills_col[i]:
        temp_df['Machine Learning'][i] = 1
    else:
         temp_df['Machine Learning'][i] = 0
    i += 1
```
